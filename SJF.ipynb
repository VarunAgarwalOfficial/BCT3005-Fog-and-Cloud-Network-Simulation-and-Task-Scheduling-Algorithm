```python
import json
```

```python
with open("data_sjf2.json" , "r") as f:
    js = json.load(f)
data = js["data"]
```

```python
hm = [0 for i in range(10)]
hm_count = [0 for i in range(10)]

for task in data:
    hm[task["priority"]] += task["total_time"]
    hm_count[task["priority"]] += 1  

avg_time = [0 for i in range(10)]
for i in range(1,10):
    avg_time[i] = hm[i] / hm_count[i]
```

```python
avg_time
```

```python
import matplotlib.pyplot as plt
from matplotlib import style


plt.style.use('ggplot')
```

```python
plt.bar( [i for i in range(1,10)], avg_time[1:])
plt.ylabel("Response Time (ms)")
plt.xlabel("Priority")
```

```python
hm2 = [0 for i in range(10)]
hm_count2 = [0 for i in range(10)]

for task in data:
    hm2[task["task_time"]] += task["total_time"]
    hm_count2[task["task_time"]] += 1  

avg_time2 = [0 for i in range(10)]
for i in range(1,10):
    avg_time2[i] = hm2[i] / hm_count2[i]
```

```python
plt.plot( [i for i in range(1,10)], avg_time2[1:])
plt.ylabel("Response Time (ms)")
plt.xlabel("Task time")
```
